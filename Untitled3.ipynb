```python
treatment_diagnosis = {
    "Insulin": ["Diabetes"],
    "Medication A": ["Hypertension"],
    "Medication B": ["Migraine"],
    "Rest & Fluids": ["Flu"],
    "No Treatment": ["Healthy"]
}
```

```python
for treatment,diagnoses in treatment_diagnosis.items():
    df["Diagnosis"] = np.where(
        df["Diagnosis"].isnull() & (df["Treatment"].isin([treatment])),
        diagnoses[0],  # Each treatment maps to one diagnosis
        df["Diagnosis"]
    )
```

```python
diagnosis_treatment = {
    "Diabetes": "Insulin",
    "Hypertension": "Medication A",
    "Migraine": "Medication B",
    "Flu": "Rest & Fluids",
    "Healthy": "No Treatment"
}
```

```python
for diagnosis, treatment in diagnosis_treatment.items():
    df["Treatment"] = np.where(
        df["Treatment"].isnull() & (df["Diagnosis"] == diagnosis),
        treatment,
        df["Treatment"])
```

```python
for treatment, diagnoses in treatment_diagnosis.items():
    print(diagnoses[0])
```

```python
gender_diagnosis = {
    "Male": ["Hypertension", "Diabetes", "Flu"],
    "Female": ["Migraine", "Healthy", "Flu"]
}
```

```python
for gender, diagnoses_list in gender_diagnosis.items():
    max_diagnosis = max(diagnoses_list)
    df["Blood_Pressure"] = np.where(
        (df["Blood_Pressure"].isnull()) & (df["Diagnosis"] == max_diagnosis),
        "120/80",  # example default value to fill
        df["Blood_Pressure"]
    )
```

```python
import pandas as pd
import plotly.express as px

# Group by 'Treatment' and calculate average 'Age'
avg_age = df.groupby('Treatment')['Age'].mean().reset_index()

# Create bar plot
fig = px.bar(avg_age,
             x='Treatment',
             y='Age',
             title='Average Age by Treatment Type',
             labels={'Treatment': 'Treatment Type', 'Age': 'Average Age'})
fig.show()
```
